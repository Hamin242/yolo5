```python
df = pd.DataFrame(yolo_val_result, columns=['File_Name'])

# '_'를 기준으로 문자열 분할
df[['model', 'aa']] = df['File_Name'].str.split('_', 1, expand=True)
df[['CAM', 'bb']] = df['aa'].str.split('_', 1, expand=True)
df[['Num', 'type']] = df['bb'].str.split('.', 1, expand=True)
df=df.drop(['aa','File_Name','bb','type'],axis=1)

for k,v in label_dic.items():
    df[v]=0
```
